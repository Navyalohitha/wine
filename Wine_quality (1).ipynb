```python
import matplotlib.pyplot as plt
plt.figure(1, figsize=(10,10))
a['quality'].value_counts().plot.pie(autopct="%1.2f%%")
```

```python
quality = a["quality"].value_counts().sort_index(ascending=True)
explode_list = [0.1, 0.1, 0, 0, 0, 0.1]
ax = quality.plot(kind='pie',
             figsize = (10,10),
             autopct='%1.1f%%',
             startangle=45,
             labels=None,
             pctdistance=1.1,
             explode=explode_list,
             textprops={'fontsize': 14})
ax.patch.set_facecolor('white')
plt.title('Wine Quality', size = 20)
ax.set(ylabel=None)
plt.axis('equal')
plt.legend(labels=quality.index, loc="upper left", fontsize=16);
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
ax = sns.distplot(a['alcohol'])
ax.set_title('Distribution of Alcohol Concentration', size=20)
ax.set(ylabel='Density', xlabel='Alcohol');
```

```python
y=a["Id"]
print(y)
```

```python
x=a.drop("Id", axis=1)
print(x)
```

```python
#LinearRegression
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

lr = LinearRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)

r2 = lr.score(X_test, y_test)
print("R^2:", r2)
```

```python
#SVM
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score, mean_squared_error
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
acc=0.245
model = SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accurary = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```
